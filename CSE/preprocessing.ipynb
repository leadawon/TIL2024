```python
import os
from pydub import AudioSegment

def split_audio(input_dir, output_dir, segment_length=15*1000):  # segment_length는 밀리초 단위
    # 입력 디렉토리에서 모든 mp3 파일을 찾습니다.
    filenames = [f for f in os.listdir(input_dir) if f.endswith('.mp3')]

    # 출력 디렉토리가 없으면 생성합니다.
    os.makedirs(output_dir, exist_ok=True)
    print(filenames[423:450])
    filenames = filenames
    blacklist = ["call_239.mp3","call_506.mp3"]
    for filename in filenames:
        if filename in blacklist:
            continue
        print(filename)
        file_path = os.path.join(input_dir, filename)
        audio = AudioSegment.from_mp3(file_path)

        
        # 파일 이름에서 확장자를 제거합니다.
        base_filename = os.path.splitext(filename)[0]

        # 오디오를 세그먼트로 나눕니다.
        for i, start in enumerate(range(0, len(audio), segment_length), 1):
            end = start + segment_length
            segment = audio[start:end]

            # 세그먼트를 파일로 저장합니다.
            segment_filename = f"{base_filename}_{i}.mp3"
            segment_path = os.path.join(output_dir, segment_filename)
            segment.export(segment_path, format="mp3")

if __name__ == "__main__":
    input_directory = "./bigdata/911_recordings"
    output_directory = "./bigdata/processed_audio"
    split_audio(input_directory, output_directory)
```
